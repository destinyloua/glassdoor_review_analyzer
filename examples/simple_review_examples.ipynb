This file contains example use cases for the SimpleReview class. 

```python
import sys 
import os
sys.path.append(os.path.abspath("../src"))

import pandas as pd

from simple_review import SimpleReview, simple_bucket

filename = "../data/all_reviews.csv"
```

## Example 1: Summarizing reviews and generating concise conclusions about employment experiences.
First, we will output the data of a review without creating a SimpleReview object. Then, we will compare the readability of the same data, but as a SimpleReview object. 

```python
# Reading the row as is from the csv file 
df_raw = pd.read_csv("../data/all_reviews.csv", nrows=1)
print("RAW DATA FROM CSV FILE: ")
print(df_raw)
```

```python
row = df_raw.iloc[0]

simple = SimpleReview(
    rating = row["rating"], 
    status = row["status"], 
    job = row["job"], 
    title = row["title"]
)
print("AFTER CONVERTING TO SimpleReview OBJECT:")
simple.display_review()
```

The first output prints the data exactly as it appears in the CSV file, while the second one is cleaner, easier to read and only contains enough fields to get a general understanding of an employee's experience. By default, the SimpleReview class will display the actual review title, but it is possible to generate different types of conclusions. 

## Example 2: Quickly understand the employee's overall happiness (impression) with the job. 
The SimpleReview class employs the strategy pattern to allow us to switch between different types of summaries. Previously, we just displayed the title of the review as the user entered. Now we can summarize the review into one word about happiness. The output below indicates the employee is generally happy with this job because they rated it more than 3 stars. In general:
    - <= 2 stars: unhappy 
    - == 3 stars: neutral 
    - > 3 stars: happy 

```python
simple.set_summary("happiness")
simple.display_review()
```

## Example 3: Quickly understand job satisfaction level. 
We can also use the SimpleReview class to determine a numerical value for job satisfaction. Unlike happiness that produces a string, the satisfaction is a percentage based on how many stars the employee rated the job (out of 5 stars). Again, we will just change the summary strategy. The output below would be interpreted as "The employee is 100% satisfied with this job."


```python
simple.set_summary("satisfaction")
simple.display_review()
```
